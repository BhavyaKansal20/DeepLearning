### Receive input signals from external sources.
### Process the inputs and decide whether to activate or not.
### Send the output signal to the next neuron or system.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("/Users/bhavyakansal/Datasets/Placement2.csv")
dataset.head(10)
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot( x = "Cgpa" , y = "Score" , data=dataset , hue="Placed")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Placed"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
x = ss.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split( x , y , test_size=0.2 , random_state=42)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr = Perceptron()
pr.fit(x_train , y_train)
```

```python
pr.score(x_train , y_train)*100 , pr.score(x_test , y_test)*100 , 
```

```python
plot_decision_regions(x , y.to_numpy() ,clf=pr)
plt.show()
```
